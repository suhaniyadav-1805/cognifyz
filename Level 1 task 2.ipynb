```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Dataset.csv")

# Display first rows
print(df.head())
```

```python
print("\nBasic Statistical Measures:\n")
print("Mean:\n", df.mean(numeric_only=True))
print("\nMedian:\n", df.median(numeric_only=True))
print("\nStandard Deviation:\n", df.std(numeric_only=True))
```

```python
country_distribution = df['Country Code'].value_counts()
print("\nCountry Code Distribution:\n", country_distribution)
```

```python
city_distribution = df['City'].value_counts()
print("\nCity Distribution:\n", city_distribution)
```

```python
cuisine_distribution = df['Cuisines'].value_counts()
print("\nCuisine Distribution:\n", cuisine_distribution)
```

```python
top_cuisines = df['Cuisines'].value_counts().head(10)
print("\nTop 10 Cuisines:\n", top_cuisines)
```

```python
top_cities = df['City'].value_counts().head(10)
print("\nTop 10 Cities:\n", top_cities)
```

```python
top_cities.plot(kind='bar', figsize=(8,5))
plt.title("Top Cities with Highest Restaurants")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
top_cuisines.plot(kind='bar', figsize=(8,5))
plt.title("Top Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
plt.hist(df['Aggregate rating'], bins=10)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```
