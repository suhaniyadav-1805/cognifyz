```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



df = pd.read_csv("Dataset.csv")


numeric_df = df.select_dtypes(include=[np.number])


print("Basic Statistical Measures using describe():")
print(numeric_df.describe())

print("\nMean:\n", numeric_df.mean())
print("\nMedian:\n", numeric_df.median())
print("\nStandard Deviation:\n", numeric_df.std())
print("\nMinimum Values:\n", numeric_df.min())
print("\nMaximum Values:\n", numeric_df.max())
```

```python
print("Country Code Distribution:")
print(df["Country Code"].value_counts())
```

```python
df["Country Code"].value_counts().plot(kind="bar")
plt.title("Distribution of Country Code")
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
print("City Distribution:")
print(df["City"].value_counts())
```

```python
top_cities = df["City"].value_counts().head(10)
print("\nTop 10 Cities with Highest Restaurants:")
print(top_cities)
```

```python

cuisine_series = df["Cuisines"].dropna()


all_cuisines = cuisine_series.str.split(", ")


from itertools import chain
flat_cuisines = list(chain.from_iterable(all_cuisines))


cuisine_count = pd.Series(flat_cuisines).value_counts()

print("Top 10 Cuisines:")
print(cuisine_count.head(10))
```

```python
cuisine_count.head(10).plot(kind="bar")
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.show()
```
