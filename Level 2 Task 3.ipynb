```python
import pandas as pd


df = pd.read_csv("Dataset.csv")
```

```python
df['Restaurant_Name_Length'] = df['Restaurant Name'].apply(len)
```

```python
df['Address_Length'] = df['Address'].apply(len)

print("New Length Features Created")
print(df[['Restaurant Name','Restaurant_Name_Length','Address_Length']].head())

```

```python
df['Has_Table_Booking'] = df['Has Table booking'].map({'Yes':1, 'No':0})
```

```python
df['Has_Online_Delivery'] = df['Has Online delivery'].map({'Yes':1, 'No':0})
```

```python
df['Number_of_Cuisines'] = df['Cuisines'].apply(lambda x: len(str(x).split(',')))
```

```python
print("\nFeature Engineered Dataset Sample:\n")

print(df[['Restaurant_Name_Length',
          'Address_Length',
          'Has_Table_Booking',
          'Has_Online_Delivery',
          'Number_of_Cuisines']].head())
```
